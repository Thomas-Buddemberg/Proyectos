### Promedios del curso

La idea de esta actividad es recibir las notas de un curso de varios alumnos de un curso para luego entregar el promedio de cada uno de ellos. En este curso hay tres pruebas $P_1$, $P_2$ y $P_3$ cuyo promedio vale un $70\%$ del curso. Además, el examen vale un $30\%$ del curso. Si la nota del examen es mayor o igual a $4.0$, entonces el promedio es el esperado. Si la nota del examen es menor que $4.0$, entonces el promedio es $min(3.9, \text{Promedio del curso})$. Asume que las notas van del 1.0 al 7.0. Finalmente, se piden algunas estadísticas sobre el curso:

- Entregar el promedio más alto y el más bajo.
- Entregar el promedio del curso.
- Entregar el número de aprobados y el número de reprobados.

Para esto, debes preguntar por el número $n$ de alumnos a ingresar. Luego debes preguntar $n$ veces por:

- El nombre del alumno.
- Las notas del alumno. Las notas serán 4 números separados por espacios, que corresponden a las 3 pruebas en orden y al examen.

Y luego entregar cada nombre de alumno, junto a su promedio. Luego debes entregar las otras estadísticas solicitadas.

```python
alumnos=[]
Nota1=[]
Nota2=[]
Nota3=[]
Examen=[]
Promedio =[]

#----------------------------------------------------------------------------------------------------------

cantidad_estudiantes = int(input('¿Cuál es la cantidad de estudiantes?: '))

if cantidad_estudiantes > 0:
    print(f'{cantidad_estudiantes} es un número de estudiantes válido')
else:
    print('Debe ingresar una cantidad de alumnos mayor a 0')
    cantidad_estudiantes = int(input('¿Cuál es la cantidad de estudiantes?: '))

for x in range(cantidad_estudiantes):
    
    nom=str(input("Ingrese el nombre del alumno: "))
    alumnos.append(nom)
    
    
    N1=float(input("Ingrese la N1 de dicho alumno: "))
    if 7>= N1 >=1:
        Nota1.append(N1)
    else:
        print('Ingrese un valor valido de nota')
        N1=float(input("Ingrese la N1 de dicho alumno: "))
        Nota1.append(N1)
    
    N2=float(input("Ingrese la N2 de dicho alumno: "))
    if 7>= N1 >=1:
        Nota2.append(N2)
    else:
        print('Ingrese un valor valido de nota')
        N2=float(input("Ingrese la N2 de dicho alumno: "))
        Nota2.append(N2)
    
    N3=float(input("Ingrese la N3 de dicho alumno: "))
    if 7>= N3 >=1:
        Nota3.append(N3)
    else:
        print('Ingrese un valor valido de nota')
        N3=float(input("Ingrese la N3 de dicho alumno: "))
        Nota3.append(N3)
    
    EX=float(input("Ingrese la nota del Examen de dicho alumno: "))
    if 7>= EX >=1:
        Examen.append(EX)
    else:
        print('Ingrese un valor valido de nota')
        EX=float(input("Ingrese la nota del Examen de dicho alumno: "))
        Examen.append(EX)
    
    if EX >= 4:
        Prom = ((N1+N2+N3)/3)*0.7 + EX*0.3
        Prom = round(Prom, 2)
        Promedio.append(Prom)
    else:
        Prom = ((N1+N2+N3)/3)*0.7 + EX*0.3
        Prom = round(Prom, 2)
        Promedio.append(min(3.9, Prom))

print('\nNotas de los alumnos:\n')
for x in range(cantidad_estudiantes):
    print(f'Nombre alumno y Promedio Final: {alumnos[x]} -> {Promedio[x]}\nNotas del alumno:\nN1: {Nota1[x]}, N2: {Nota2[x]}, N3: {Nota3[x]}, Examen: {Examen[x]}\n')
    
#----------------------------------------------------------------------------------------------------------------------

alto = max(Promedio)

bajo = min(Promedio)

print(f'Promedio más alto: {alto}\nPromedio más bajo: {bajo}')

#----------------------------------------------------------------------------------------------------------------------

suma = 0

for x in range(cantidad_estudiantes):
    suma = suma + Promedio[x]
    
Promedio_curso = suma/cantidad_estudiantes

print(f'\nPromedio del curso: {round(Promedio_curso, 3)}')

#----------------------------------------------------------------------------------------------------------------------

Aprobados = []
Reprobados = []

for x in range(cantidad_estudiantes):
    if Promedio[x] >= 4:
        Aprobados.append(Promedio[x])
    else:
        Reprobados.append(Promedio[x])

print(f'\nNúmero de alumnos Aprobados: {len(Aprobados)}\nNúmero de alumnos Reprobados: {len(Reprobados)}')
```
