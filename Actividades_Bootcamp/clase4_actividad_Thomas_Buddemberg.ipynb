La primera se llama Queue (o cola).
Esta es:
- Secuencial
- Finita
- Flexible
- Se encuentra bajo el principio FIFO: First in First Out

FIFO: Esto significa que el primer elemento de la Cola es el primer en salir.

Ejemplo:
1,2,3,4
Se saca un elemento
2,3,4

Bajo esta lógica debera crear la estrucutra con 2 funciones Dequeue y Enqueue

Dequeue: Esta función es la encargada de eliminar un dato de la lista

Ejemplo:
1,2,3,4
Se saca un elemento
2,3,4

Enqueue: Esta función esta encargada de agregar un dato a la lista

Ejemplo:
1,2,3,4
Se agrega el elemento 5
1,2,3,4,5

```python
class CNode:
    def __init__(self,valor):
        self.valor = valor
        self.nodo_siguiente = None

    def __str__(self):
        return str(self.valor)
    
    def __repr__(self):
        return str(self.valor)
    
#--------------------------------------------------------------------------------------------------------------

class Queue:
    def __init__(self):
        self.cabeza = None 
        self.cola = None
        self.tamaño = 0

    def __str__(self):
        lista = []
        nodo_actual = self.cabeza
        pivote = True
        contador = self.tamaño
        while contador != 0:
            if pivote != False or nodo_actual != self.cabeza:
                lista.append(nodo_actual.valor)
                nodo_actual = nodo_actual.nodo_siguiente
                pivote = False
                contador -= 1
            else:
                break
        return str(lista) + " Tamaño: " + str(self.tamaño)
    
    def enqueue(self, valor):
      nuevo_nodo = CNode(valor)
      if self.cabeza == None and self.cola == None:
          self.cabeza = nuevo_nodo
          self.cola = nuevo_nodo
      else:
          nuevo_nodo.nodo_siguiente = self.cabeza
          self.cola.nodo_siguiente = nuevo_nodo
          self.cabeza = nuevo_nodo
      self.tamaño += 1

      return self

    def dequeue(self, valor):
        nuevo_nodo = CNode(valor)
        if self.cabeza == None and self.cola == None:
            self.cabeza = nuevo_nodo
            self.cola = nuevo_nodo
        else:
            self.cola.nodo_siguiente = nuevo_nodo
            nuevo_nodo.nodo_siguiente = self.cabeza
            self.cola = nuevo_nodo
        self.tamaño += 1

        return self
```

```python
lista = Queue()
#Enqueue: Esta función esta encargada de agregar un dato a la lista
lista.enqueue(8)
print(lista)

#Dequeue: Esta función es la encargada de eliminar un dato de la lista
lista.dequeue(2)
print(lista)
```
